Some details about the dataset and columns 

```python
price = pd.read_csv("house_price.csv")
price.head()
```

```python
price.info(memory_usage = "deep") #memory usage and also datatypes of all columns
```

```python
price.describe().round() #from this we could check which values can be downcast
```

```python
price.isna().sum() #checking if there is any missing values
```

```python
price["city"].nunique() #total number of unique cities
```

```python
price["floors"].nunique() #total number of floors
```

```python
price["sqft_living"].max() #house with maximum size for living rooms in sqft
```

```python
price["price"].mean() #average price of houses
```

```python
price.loc[price["price"].argmax()] #all details about the house with maximum price
```

```python
price.loc[price["price"].argmin()] #all details about the house with minimum price
```

```python
price["sqft_basement"].value_counts()[0] #housees without basement such as there is no space for basement
```

```python
price.loc[price["sqft_above"].argmax()] #the house wchich has maximum space on rooftop
```

```python
city_price = price.groupby("city")["price"].mean().sort_values(ascending = False) #average price for houses in per city
```

```python
city_price
```

```python
import matplotlib.pyplot as plt
```

```python
#plot as bar chart
plt.figure(figsize=(12,6))
city_price.plot(kind='bar', color='skyblue')
plt.title("Average house price by city")
plt.xlabel("City")
plt.ylabel("Average Price ($)")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
